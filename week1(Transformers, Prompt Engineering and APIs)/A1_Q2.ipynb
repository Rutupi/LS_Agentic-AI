```python
# ...existing code...
from openai import OpenAI
import os

# Read key from env var (recommended). Raise helpful error if missing.
api_key = os.environ.get("OPENAI_API_KEY")
if not api_key:
    raise EnvironmentError(
        "OPENAI_API_KEY not set. Set it in your OS environment or pass a key to OpenAI() (not recommended)."
    )

client = OpenAI(api_key=api_key)

response = client.responses.create(
    model="gpt-5.5",  # make sure this model is available on your account; change if needed
    input="Write a one-sentence bedtime story about a unicorn.",
)

print(response.output_text)
# ...existing code...
```
