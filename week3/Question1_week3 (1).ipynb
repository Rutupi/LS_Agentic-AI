```python
# Instead of langchain.agents, use langchain_classic.agents
from langchain_classic.agents import AgentExecutor, create_react_agent
# Keep community and core tools as they are
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_core.tools import tool
# Import hub from the standalone langchainhub package

from langchain_openai import ChatOpenAI
import langchainhub
# 2. Define our custom calculator tool
@tool
def calculator(expression: str) -> str:
    """Useful for answering math questions. Input should be a valid mathematical expression."""
    try:
        # Note: eval() is used here for simplicity in a controlled assignment environment.
        return str(eval(expression))
    except Exception as e:
        return f"Error evaluating expression: {e}"

def run_react_agent():
    # 3. Initialize tools and the LLM
    search = DuckDuckGoSearchRun()
    tools = [search, calculator]
    llm = ChatOpenAI(temperature=0, model="gpt-4o-mini")

    # 4. Pull the standard ReAct prompt from the LangChain hub
prompt = hub.pull("hwchase17/react", dangerously_pull_public_prompt=True)
    # 5. Construct the ReAct agent and the executor using the classic imports
    agent = create_react_agent(llm, tools, prompt)
    agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True, handle_parsing_errors=True)

    # 6. Define 3 queries that force the use of both tools
    queries = [
        "What is the current population of Canada?", # Forces Search
        "What is 456 multiplied by 987?", # Forces Calculator
        "Find the height of the Eiffel Tower in meters and multiply it by 15." # Forces Both
    ]

    for q in queries:
        print(f"\n{'='*50}\nRunning query: {q}\n{'='*50}")
        # The verbose=True flag in the executor will print the Thought -> Action -> Observation trace
        agent_executor.invoke({"input": q})

if __name__ == "__main__":
    run_react_agent()
```
